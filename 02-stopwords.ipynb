```python
df = pd.read_csv("C:/Curso-ML-1/project/clean-test-cross.csv")
```

```python
## removing the stop words of all colunms
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize
stop_words = set(stopwords.words('english')) 
df2 = df  
df2["selftext_clean"] =''
df2["selftext_clean1"] =''
len1 = len(df)
for i in range(len1):  
    example_sent1 =  str(df["articleBody"][i])
    example_sent2 =  str(df["Headline"][i])


    
  
    word_tokens1 = word_tokenize(example_sent1) 
    word_tokens2 = word_tokenize(example_sent2)

    

                 
    filtered_sentence1 = [w for w in word_tokens1 if not w in stop_words] 
    filtered_sentence1 = [] 
    filtered_sentence2 = [w for w in word_tokens2 if not w in stop_words] 
    filtered_sentence2 = [] 
   
    
   
    for w in word_tokens1: 
        if w not in stop_words: 
            filtered_sentence1.append(w) 
    df2["selftext_clean1"][i] = filtered_sentence1
    for w in word_tokens2: 
        if w not in stop_words: 
            filtered_sentence2.append(w) 
    df2["selftext_clean"][i] = filtered_sentence2
 
 
    

    ##print(filtered_sentence1) 
    ##print(filtered_sentence2) 

    

df2.to_csv('C:/Curso-ML-1/project/nostopwords-test-cross.csv')
```
