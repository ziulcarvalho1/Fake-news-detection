```python
df = pd.read_csv("C:/Curso-ML-1/project/nostopwords-test-cross.csv")
```

```python
df["selftext_lemma"] =''
df["selftext_nouns"] = ''
df["selftext_adjectives"] = ''
df["selftext_verbs"] = ''
df["selftext_nav"] =''
df["no_tokens"] = 0

df["selftext_lemma1"] =''
df["selftext_nouns1"] = ''
df["selftext_adjectives1"] = ''
df["selftext_verbs1"] = ''
df["selftext_nav1"] =''
df["no_tokens1"] = 0
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
for i, row in df.iterrows():
    if i % 100 == 0:
        print(i)
    if(row["selftext_clean"] and len(str(row["selftext_clean"])) < 200000):
        doc = nlp(str(row["selftext_clean"]))
        adjectives = []
        nouns = []
        verbs = []
        lemmas = []

        for token in doc:
            lemmas.append(token.lemma_)
            if token.pos_ == "ADJ":
                adjectives.append(token.lemma_)
            if token.pos_ == "NOUN" or token.pos_ == "PROPN":
                nouns.append(token.lemma_)
            if token.pos_ == "VERB":
                verbs.append(token.lemma_)
                
        df.at[i, "selftext_lemma"] = " ".join(lemmas)                
        df.at[i, "selftext_nouns"] = " ".join(nouns)
        df.at[i, "selftext_adjectives"] = " ".join(adjectives)
        df.at[i, "selftext_verbs"] = " ".join(verbs)
        df.at[i, "selftext_nav"] = " ".join(nouns+adjectives+verbs)
        df.at[i, "no_tokens"] = len(lemmas)
```

```python
for i, row in df.iterrows():
    if i % 100 == 0:
        print(i)
    if(row["selftext_clean1"] and len(str(row["selftext_clean1"])) < 200000):
        doc = nlp(str(row["selftext_clean1"]))
        adjectives = []
        nouns = []
        verbs = []
        lemmas = []

        for token in doc:
            lemmas.append(token.lemma_)
            if token.pos_ == "ADJ":
                adjectives.append(token.lemma_)
            if token.pos_ == "NOUN" or token.pos_ == "PROPN":
                nouns.append(token.lemma_)
            if token.pos_ == "VERB":
                verbs.append(token.lemma_)
                
        df.at[i, "selftext_lemma1"] = " ".join(lemmas)                
        df.at[i, "selftext_nouns1"] = " ".join(nouns)
        df.at[i, "selftext_adjectives1"] = " ".join(adjectives)
        df.at[i, "selftext_verbs1"] = " ".join(verbs)
        df.at[i, "selftext_nav1"] = " ".join(nouns+adjectives+verbs)
        df.at[i, "no_tokens1"] = len(lemmas)
```

```python
df.to_csv('C:/Curso-ML-1/project/separated-test-cross.csv')
```
