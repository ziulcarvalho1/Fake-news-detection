```python
## Calculates the TF-IDF of the BOW file - This file has the BOW values, One line with the words, One line with the number of documets
## and subsequentment the values TF-IDF tail(Len(df))
import math
df = pd.read_csv("C:/Curso-ML-1/project/TF-IDF-headline.csv")
###df =df.head(5)
df.info()

len3 = len(df)

print(df)
for i in range(len3):
    x=[]
    x.append(" ")
    x.append(" ")
    x.append(" ")
    for ix in range(199):
        if int(df["200"][i])!=0:
            tf = int(df[str(ix)][i])/int(df["200"][i])
        DF = len3/int(df[str(ix)][len3-1])+1
        factor = tf*math.log(DF)
        ##print(factor)
        x.append(str(factor))
    print(x)
    x.append(" ")
    x.append(" ")
    to_append = x
    a_series = pd. Series(to_append, index = df.columns)
    df = df.append(a_series, ignore_index=True)   

print(df)

df.to_csv('C:/Curso-ML-1/project/TF-IDF-headlinex.csv')
```
