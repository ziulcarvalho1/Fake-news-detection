```python
# Import
import requests
import nltk
import urllib.request
import time
import pandas, numpy
import pandas as pd 
import numpy as np

from bs4 import BeautifulSoup
from newspaper import Article
nltk.download('punkt')
```

```python
#Preparing the file to store the urls
df1 = pd.read_csv("C:/Curso-ML-1/project/source-nyt.csv") 
##df =df.head(10000)
df1.info()
df1['text1']=""

df = pd.read_csv("C:/Curso-ML-1/project/Pasta1.csv")
print(df)
df.info()
df['headline']= ""
df['text']    = ""
df['date']    = ""
df['author']  = ""
df['source']  = ""
df['fake']    = ""
                 
df.info()
```

```python
len1 = len(df1)
for i in range(len1):
    len2 = len(df1['0'][i])
    url3 = df1['0'][i]
    selector="ON"
    for ix in range(len2):
        caract = url3[ix]
        if caract=='2' and selector=="ON":
            print(i)
            selector = "OFF"
            df1['text1'][i] = 'https://www.nytimes.com/'+url3[ix:]
            print(df1['text1'][i])
```

```python
print(df1['text1'])
```

```python
len1 = len(df1)
for i in range(95,len1):
    
    
    URL1 = df1['text1'][i]
    print(URL1)
    print(i)
    article = Article(URL1)
    article.download()
    article.parse()
    article.nlp()
    
    author1   = article.authors
    date1     = article.publish_date
    text1     = article.text
    headline1 = article.title
    source1   = "Reuters"
    fake1     = "NO"
    
    article2 = [[" ",headline1, text1, date1 , author1,source1,fake1]]
    ##print(article2)
    df3 = pd.DataFrame(article2, columns = ["   ",'headline' , 'text', 'date' , 'author','source','fake'])
    df = df.append(df3, ignore_index=True)
```
