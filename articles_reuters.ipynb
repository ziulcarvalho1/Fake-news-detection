# Loading articles

```python
# Import
import requests
import nltk
import urllib.request
import time
import pandas, numpy
import pandas as pd 
import numpy as np

from bs4 import BeautifulSoup
from newspaper import Article
nltk.download('punkt')
```

```python
#Preparing the file to store the urls
df1 = pd.read_csv("C:/Curso-ML-1/project/source-reuters.csv") 
##df =df.head(10000)
df1.info()

df = pd.read_csv("C:/Curso-ML-1/project/Pasta1.csv")
print(df)
df.info()
df['headline']= ""
df['text']    = ""
df['date']    = ""
df['author']  = ""
df['source']  = ""
df['fake']    = ""
                 
df.info()
```

```python
# cleaning the source
len2 = len(df1)
for i in range(len2):
    correct =  df1['0'][i]
    ##print(i)
    ##print(correct[len(correct)-16:len(correct)-15])
    if (correct[len(correct)-16:len(correct)-15])=='"':
        print(i)
        print(df1['0'][i])
        df1['0'][i]=correct[0:len(correct)-16]

```

```python
df1['0'][140]="https://www.reuters.com/places/brazil/article/us-reuters-editorial-leadership/reuters-editorial-leadership-idUSKBN1KE2SD"
```

```python
len1 = len(df1)
for i in range(141,len1):
    URL1 = df1['0'][i]
    print(URL1)
    print(i)
    article = Article(URL1)
    article.download()
    article.parse()
    article.nlp()
    
    author1   = article.authors
    date1     = article.publish_date
    text1     = article.text
    headline1 = article.title
    source1   = "Reuters"
    fake1     = "NO"
    
    article2 = [[" ",headline1, text1, date1 , author1,source1,fake1]]
    ##print(article2)
    df3 = pd.DataFrame(article2, columns = ["   ",'headline' , 'text', 'date' , 'author','source','fake'])
    df = df.append(df3, ignore_index=True)
```

```python
df.to_csv('C:/Curso-ML-1/project/articles-reuters.csv')
```
